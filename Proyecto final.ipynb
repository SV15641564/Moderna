```python
#ADAPTADO DE OPENAI
import numpy as np
import matplotlib.pyplot as plt
from scipy.constants import hbar, electron_volt

def probabilidad_penetracion(m, d, V0, E):
    #k1=alpha
    k1 = np.sqrt(2 * m * (V0 - E)) / hbar
    T = 1 / (1 + (V0**2 * (np.sinh(k1 * d))**2) / (4 * E * (V0 - E)))
    return T

# Parámetros de la barrera de potencial y la partícula
m = 9.10938356e-31  # masa del electrón en kg
d = 1e-9  # anchura de la barrera en metros
V0 = 10 * electron_volt  # altura de la barrera en eV

# Rango de energías de enlace
Energias = np.linspace(0, 10, 1000) * electron_volt

# Calcular la probabilidad de penetración para cada energía de enlace
Probabilidades = [probabilidad_penetracion(m, d, V0, E) for E in Energias]

# Graficar los resultados
plt.plot(Energias / electron_volt, Probabilidades)
plt.xlabel("Energía de enlace (eV)")
plt.ylabel("Probabilidad de penetración")
plt.title("Probabilidad de penetración vs. Energía de enlace")
plt.grid()
plt.show()
```

```python
#ADAPTADO DE OPENAI
import numpy as np
import matplotlib.pyplot as plt
from scipy.constants import hbar, electron_volt

def probabilidad_penetracion(m, a, V0, E):
    k1 = np.sqrt(2 * m * E) / hbar
    k2 = np.sqrt(2 * m * (V0 - E)) / hbar
    T = np.exp(-2*k2*a)
    return T

# Parámetros de la barrera de potencial y la partícula
m = 9.10938356e-31  # masa del electrón en kg
a = 1e-10  # anchura de la barrera en metros
V0 = 10 * electron_volt  # altura de la barrera en eV

# Rango de energías de enlace
Energias = np.linspace(0, 10, 1000) * electron_volt

# Calcular la probabilidad de penetración para cada energía de enlace
Probabilidades = [probabilidad_penetracion(m, a, V0, E) for E in Energias]

# Graficar los resultados
plt.plot(Energias / electron_volt, Probabilidades)
plt.xlabel("Energía de enlace (eV)")
plt.ylabel("Probabilidad de penetración")
plt.title("Probabilidad de penetración vs. Energía de enlace")
plt.grid()
plt.show()
```

#BINDING ENERGY DE DIFERENTES ELEMENTOS

valores teóricos vs experimentales
image.png

https://en.wikipedia.org/wiki/Semi-empirical_mass_formula

1.	"volume coefficient nuclear binding energy" para a_v.
2.	"surface coefficient nuclear binding energy" para a_s.
3.	"Coulomb coefficient nuclear binding energy" para a_c.
4.	"asymmetry coefficient nuclear binding energy" para a_a.


```python
#ADAPTADO DE OPENAI
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score


def binding_energy(Z, A):
    a_v = 15.75
    a_s = 17.8
    a_c = 0.711
    a_a = 23.7
    a_p= 12.0

    if A % 2 == 1:
        delt = 0
    elif Z % 2 == 0:
        delt = a_p/(A**0.5)
    else:
        delt = -(a_p/(A**0.5))

    B = ((a_v * A) - (a_s * A**(2/3)) - (a_c * Z*(Z-1)) / A**(1/3)) - (a_a * (A - 2 * Z)**2 / A) + delt 
    return B

def plot_binding_energy(experimental, theoretical, elements):
    plt.plot(elements, experimental, 'o', label='Experimental')
    plt.plot(elements, theoretical, 'x', label='Teórico')

    plt.xlabel('Elementos')
    plt.ylabel('Energía de enlace (MeV)')
    plt.title('Energía de enlace: Experimental vs. Teórico')

    for i, element in enumerate(elements):
        plt.text(element, experimental[i], element, ha='center', va='bottom')

    r2 = r2_score(experimental, theoretical)
    plt.text(0.95, 0.05, f'R^2 = {r2:.3f}', transform=plt.gca().transAxes, ha='right', va='bottom')

    plt.legend()
    plt.show()

experimental_data = [2.22, 28.30, 39.24, 58.16, 92.16]
ZA_values = [(1, 2), (2, 4), (3, 7), (4, 9), (6, 12)]
elements = ['H-2', 'He-4', 'Li-7', 'Be-9', 'C-12']

theoretical_data = [binding_energy(Z, A) for Z, A in ZA_values]

plot_binding_energy(experimental_data, theoretical_data, elements)
```
